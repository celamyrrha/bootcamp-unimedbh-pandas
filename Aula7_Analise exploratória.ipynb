```python
#criando o dataframe
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
#visualizando as 5 primeiras linhas
df.head()
```

```python
#Quantidade de linhas e colunas
df.shape
```

```python
#Verificando os tipos de dados
df.dtypes
```

```python
#Qual a receita total?
df["Valor Venda"].sum()
```

```python
#Qual o custo total?
df["custo"]= df["Custo Unitário"].mul(df["Quantidade"])#Criando a coluna de custo
```

```python
#qual o custo total?
round(df["custo"].sum(),2)
```

```python
#Criar coluna de lucro Receita - custo
df["lucro"] = df["Valor Venda"] - df["custo"]
```

```python
#qual o lucro total?
round(df["lucro"].sum(),2)
```

```python
#Criando uma coluna para o tempo de envio
df["tempo_envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head()
```

```python
#Passando para numérico
df["tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df["tempo_envio"].dtypes
```

```python
#média do tempo de envio por marca
df.groupby("Marca")["tempo_envio"].mean()
```

```python
#verificando se tempo dados faltantes
df.isnull().sum()
```

```python
#vamos agrupar por ano e marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#resetando o index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum().reset_index()
lucro_ano
```

```python
#Qual o total de produtos vendidos?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending = False)
```

```python
#Gráfico total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending = True).plot.barh(title="Total de produtos vendidos")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
df.groupby(df["Data Venda"].dt.year)["lucro"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Lucro");
```

```python
#selecionando as vendas do ano de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head()
```

```python
df_2009.groupby(df_2009["Data Venda"].dt.month)["lucro"].sum().plot(title="Lucro x mês")
plt.xlabel("mês")
plt.ylabel("lucro");
```

```python
df_2009.groupby(df_2009["Marca"])["lucro"].sum().plot.bar(title="Lucro x marca")
plt.xlabel("marca")
plt.ylabel("lucro")
plt.xticks(rotation="horizontal");
```

```python
df_2009.groupby(df_2009["Classe"])["lucro"].sum().plot.bar(title="Lucro x Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation="horizontal");
```
